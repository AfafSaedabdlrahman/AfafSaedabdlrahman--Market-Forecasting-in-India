#  problem formulation

* **The problem**  

Analyzing and predicting the behavior of a cellular network over 15 consecutive days using a dataset containing cell performance data. The goal is to forecast two key features, feature_11 and feature_12, for each cell on a given day, based on the features of that day and the preceding day, using the geographical position of the cells and various performance-related features.
Input:

    Dataset: A 15-day dataset representing the performance of cells in a network, where each day's data consists of 23 columns (features) and 171 rows (representing individual cells).
        Column 1: Cell identifier, which links data across the 15 days.
        Columns 2-23: Cell features, including spatial positions (cell_x, cell_y, cell_z) and performance metrics such as feature_11 and feature_12.
        Sequential Data: The 15 days represent the changing states of the environment with different actions impacting the network's performance.

Output:

    Prediction: A model that forecasts feature_11 and feature_12 for each cell on day t+1 using all other features from day t and day t+1, except feature_11 and feature_12.
    Evaluation Metrics: Metrics such as Mean Squared Error (MSE) or Root Mean Squared Error (RMSE) to assess the predictive performance of the model.
* **The challenges**  
* Time Dependency: The data is sequential, meaning temporal relationships between days must be captured. The model must understand how actions taken one day influence the next.
* Feature Relationships: The model should identify relationships between feature_11, feature_12, and other features, not just on a single day but across consecutive days.
* Geospatial Data: Incorporating geographical information from cell_x, cell_y, and cell_z into the model to improve predictive accuracy. Finding a suitable data representation (e.g., distance metrics, clustering) for spatial features could be complex.
Data Complexity: Handling a relatively small dataset (171 cells) over 15 days may lead to overfitting, especially with multiple features to analyze.
* Feature Exclusion: The model must predict feature_11 and feature_12 without knowing their values for the next day, which increases the complexity of the prediction task.


* **The impact of the problem** 
** Improved Network Performance: By accurately predicting performance metrics (such as feature_11 and feature_12), network operators can proactively optimize network performance and take preventive actions.
** Resource Allocation: Insights from this predictive model can guide optimal resource allocation (e.g., adjusting bandwidth, increasing power) in areas where performance is predicted to decline.

```python
def calculate_spearman_correlations(df):
    """
    Calculates Spearman's correlation for all numerical column combinations in a DataFrame.

    Args:
        df: The input DataFrame.

    Returns:
        A pandas DataFrame containing correlation, p-value, and significance for each pair of columns.
    """

    numeric_cols = df.select_dtypes(include=np.number).columns
    combinations = list(itertools.combinations(numeric_cols, 2))

    results = []
    for c1, c2 in combinations:
        correlation, p_value = stats.spearmanr(df[c1], df[c2])
        alpha = 0.05
        significant = p_value < alpha
        results.append([c1, c2, correlation, p_value, significant])

    correlation_df = pd.DataFrame(results, columns=['Column1', 'Column2', 'Correlation', 'p_value', 'Significant'])
    return correlation_df

calculate_spearman_correlations(df)
```

```python
def mannwhitneyu_loop(df, target_columns):
    """
    Performs Mann-Whitney U test for all numerical features against two target columns.

    Args:
        df: The input DataFrame.
        target_columns: A list of two target columns.

    Returns:
        A pandas DataFrame containing test statistics, p-values, and significance for each feature.
    """

    numeric_cols = df.select_dtypes(include=np.number).columns
    results = []

    for feature in numeric_cols:
        if feature not in target_columns:
            for target_column in target_columns:
                # Use median to split the target column into two groups
                median_value = df[target_column].median()
                group_low = df[df[target_column] <= median_value][feature]
                group_high = df[df[target_column] > median_value][feature]

                # Perform Mann-Whitney U test
                stat, p_value = stats.mannwhitneyu(group_low, group_high)
                alpha = 0.05
                significant = p_value < alpha
                results.append([feature, target_column, stat, p_value, significant])

    result_df = pd.DataFrame(results, columns=['Feature', 'Target', 'Statistic', 'p_value', 'Significant'])
    return result_df


mannwhitneyu_loop(df, ['feature_11_int', 'feature_12'])
```

```python
from scipy.stats import pearsonr
import pandas as pd

def pearson_correlation_test_with_pvalues(df):
    """
    Performs Pearson correlation test between all numerical features to check for linearity,
    including both the correlation coefficient and the associated p-value.

    Args:
        df: The input DataFrame.

    Returns:
        A pandas DataFrame containing correlation coefficients, p-values, and significance for each feature pair.
    """
    
    numeric_cols = df.select_dtypes(include=np.number).columns
    results = []
    
    # Calculate the Pearson correlation and p-value for each pair of numerical features
    for i, col1 in enumerate(numeric_cols):
        for col2 in numeric_cols[i+1:]:
            corr, p_value = pearsonr(df[col1], df[col2])
            alpha = 0.05
            significant = p_value <= alpha
            results.append([col1, col2, corr, p_value, significant])
    
    result_df = pd.DataFrame(results, columns=['Column1', 'Column2', 'Correlation', 'p-value', 'Significant'])
    return result_df


pearson_correlation_test_with_pvalues(df)
```

```python


def prepare_data(df, non_feature_columns):
    """
    Prepares the dataset for time series forecasting using sliding windows.
    
    Parameters:
        df (pd.DataFrame): The original dataframe with features.
        non_feature_columns (list): List of columns to drop from the dataset.
    
    Returns:
        X (np.ndarray): Array containing the input features for training.
        y (np.ndarray): Array containing the target output for training.
    """
    # Drop non-feature columns to focus on relevant features
    forecast_df_features = df.drop(columns=non_feature_columns).copy()

    X_list = []
    y_list = []

    # Create sliding windows over the data
    for i in range(len(forecast_df_features) - 1):
        # Current day features (excluding target features)
        current_day_features = forecast_df_features.iloc[i].drop(['feature_11_int', 'feature_12']).values
        
        # Next day features (excluding target features)
        next_day_features = forecast_df_features.iloc[i + 1].drop(['feature_11_int', 'feature_12']).values
        
        # Combine features from the current and next day
        window_features = np.concatenate([current_day_features, next_day_features])
        
        # Target output: feature_11_int and feature_12 of the next day
        target_output = forecast_df_features.iloc[i + 1][['feature_11_int', 'feature_12']].values
        
        # Append the current window's features and target to lists
        X_list.append(window_features)
        y_list.append(target_output)

    # Convert lists to numpy arrays for model training
    X = np.array(X_list)
    y = np.array(y_list)
    
    return X, y
```

```python
from xgboost import XGBRegressor

def train_model(X_train, y_train):
    """
    Trains an XGBRegressor model.
    
    Parameters:
        X_train (np.ndarray): Training feature data.
        y_train (np.ndarray): Training target data.
    
    Returns:
        model: Trained XGBRegressor model.
    """
    # Initialize the XGBRegressor model
    model = XGBRegressor(n_estimators=1000, learning_rate=0.1, random_state=42)
    
    # Fit the model to the training data
    model.fit(X_train, y_train)
    
    return model
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

def evaluate_model(model, X_test, y_test):
    """
    Evaluates the model using Mean Squared Error (MSE) and Root Mean Squared Error (RMSE).
    
    Parameters:
        model: Trained XGBRegressor model.
        X_test (np.ndarray): Test feature data.
        y_test (np.ndarray): Test target data.
    
    Returns:
        None
    """
    # Predict the target values using the trained model
    y_pred = model.predict(X_test)

    # Calculate Mean Squared Error and Root Mean Squared Error
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    print(f"Mean Squared Error: {mse:.4f}")
    print(f"Root Mean Squared Error: {rmse:.4f}")

    # Evaluate individual target features separately
    mse_feature_11 = mean_squared_error(y_test[:, 0], y_pred[:, 0])
    mse_feature_12 = mean_squared_error(y_test[:, 1], y_pred[:, 1])
    rmse_feature_11 = np.sqrt(mse_feature_11)
    rmse_feature_12 = np.sqrt(mse_feature_12)
    
    print(f"Mean Squared Error for feature_11_int: {mse_feature_11:.4f}")
    print(f"Root Mean Squared Error for feature_11_int: {rmse_feature_11:.4f}")
    print(f"Mean Squared Error for feature_12: {mse_feature_12:.4f}")
    print(f"Root Mean Squared Error for feature_12: {rmse_feature_12:.4f}")
```

```python

# Function to apply spatial clustering (e.g., K-Means) to geographical data
def apply_spatial_clustering(df, n_clusters=5):
    """
    Applies K-Means clustering to the spatial coordinates (cell_x, cell_y, cell_z) of the cells.

    Parameters:
        df (pd.DataFrame): DataFrame containing spatial columns (cell_x, cell_y, cell_z).
        n_clusters (int): Number of clusters for K-Means.
    
    Returns:
        pd.DataFrame: DataFrame with an additional column for cluster labels.
    """
    # Extract spatial coordinates
    spatial_data = df[['cell_x', 'cell_y', 'cell_z']]
    
    # Standardize the spatial data
    scaler = StandardScaler()
    spatial_data_scaled = scaler.fit_transform(spatial_data)
    
    # Apply K-Means clustering
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    df['spatial_cluster'] = kmeans.fit_predict(spatial_data_scaled)
    
    return df

# Function to calculate distance-based features
def calculate_distance_features(df):
    """
    Calculates distance-based features, such as distance from the origin (0,0,0).

    Parameters:
        df (pd.DataFrame): DataFrame containing spatial columns (cell_x, cell_y, cell_z).
    
    Returns:
        pd.DataFrame: DataFrame with additional columns for distance-based features.
    """
    # Calculate the Euclidean distance from the origin (0, 0, 0)
    df['distance_from_origin'] = np.sqrt(df['cell_x']**2 + df['cell_y']**2 + df['cell_z']**2)
    
    # Calculate distances from the centroid (mean of all points)
    centroid = df[['cell_x', 'cell_y', 'cell_z']].mean().values
    df['distance_from_centroid'] = np.sqrt(((df[['cell_x', 'cell_y', 'cell_z']] - centroid) ** 2).sum(axis=1))
    
    return df

# Function to prepare the sliding windows for training data, updated with spatial features
def prepare_data_with_spatial(df, non_feature_columns):
    """
    Prepares the dataset for time series forecasting using sliding windows.
    Updates data with spatial features (clustering and distances).

    Parameters:
        df (pd.DataFrame): The original dataframe with features.
        non_feature_columns (list): List of columns to drop from the dataset.
    
    Returns:
        X (np.ndarray): Array containing the input features for training.
        y (np.ndarray): Array containing the target output for training.
    """
    # Apply spatial clustering and distance features
    df = apply_spatial_clustering(df, n_clusters=5)
    df = calculate_distance_features(df)

    # Drop the non-feature columns to get only relevant features for modeling
    forecast_df_features = df.drop(columns=non_feature_columns).copy()

    print(forecast_df_features.head())  # Check if the features are prepared correctly

    # Initialize lists to store the sliding window data
    X_list, y_list = [], []

    # Iterate through the dataframe to create sliding windows
    for i in range(len(forecast_df_features) - 1):
        # Current day features (excluding 'feature_11_int' and 'feature_12')
        current_day_features = forecast_df_features.iloc[i].drop(['feature_11_int', 'feature_12']).values
        
        # Next day features (excluding 'feature_11_int' and 'feature_12')
        next_day_features = forecast_df_features.iloc[i + 1].drop(['feature_11_int', 'feature_12']).values
        
        # Combine the current and next day's features
        window_features = np.concatenate([current_day_features, next_day_features])
        
        # Target output: 'feature_11_int' and 'feature_12' of the next day
        target_output = forecast_df_features.iloc[i + 1][['feature_11_int', 'feature_12']].values
        
        # Append the data to the lists
        X_list.append(window_features)
        y_list.append(target_output)

    # Convert the lists into numpy arrays
    X_array, y_array = np.array(X_list), np.array(y_list)
    
    print(f"X shape: {X_array.shape}, y shape: {y_array.shape}")  # Check the shapes of X and y
    
    return X_array, y_array

# Function to train and evaluate model with hyperparameter tuning
def train_and_evaluate_model(X_train, X_test, y_train, y_test, feature_names):
    """
    Trains and evaluates a model using XGBRegressor and performs hyperparameter tuning.

    Parameters:
        X_train (np.ndarray): Training feature data.
        X_test (np.ndarray): Test feature data.
        y_train (np.ndarray): Training target data.
        y_test (np.ndarray): Test target data.
        feature_names (list): List of feature names for plotting.
    
    Returns:
        None
    """
    # Initialize XGBRegressor
    model = XGBRegressor(n_estimators=1000, learning_rate=0.1, random_state=42)

    # Train the model
    model.fit(X_train, y_train)

    # Predict on test set
    y_pred = model.predict(X_test)

    # Evaluate the model
    mse = mean_squared_error(y_test, y_pred)
    print(f"Mean Squared Error: {mse:.4f}")

    # Evaluate individual target features (feature_11_int and feature_12) separately
    mse_feature_11 = mean_squared_error(y_test[:, 0], y_pred[:, 0])
    mse_feature_12 = mean_squared_error(y_test[:, 1], y_pred[:, 1])
    print(f"Mean Squared Error for feature_11_int: {mse_feature_11:.4f}")
    print(f"Mean Squared Error for feature_12: {mse_feature_12:.4f}")

    # Plot the predictions vs true values
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.scatter(y_test[:, 0], y_pred[:, 0], alpha=0.5)
    plt.title("Predictions vs True values for feature_11_int")
    plt.xlabel("True values")
    plt.ylabel("Predictions")

    plt.subplot(1, 2, 2)
    plt.scatter(y_test[:, 1], y_pred[:, 1], alpha=0.5)
    plt.title("Predictions vs True values for feature_12")
    plt.xlabel("True values")
    plt.ylabel("Predictions")

    plt.show()

# Main function to run the entire pipeline
def main(df):
    """
    Main function to execute the entire modeling pipeline with spatial features.

    Parameters:
        df (pd.DataFrame): The input dataframe with features and target values.
    """
    # Define the columns to drop (non-features)
    non_feature_columns = ['cell_x', 'cell_y', 'cell_z', 'feature_11', 'feature_9']
    
    # Prepare the sliding window data with spatial features
    X, y = prepare_data_with_spatial(df, non_feature_columns)
    
    # Split the data into training and test sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True, random_state=42)

    # Train the model and evaluate performance with feature importance
    print("Training model and evaluating feature importance...")
    train_and_evaluate_model(X_train, X_test, y_train, y_test, feature_names=[])


```
